```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from csv import reader
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data=pd.read_csv("store_data.txt",header=None)
data.head()
```

```python
products = []
with open('store_data.txt', 'r') as read_obj:
    csv_reader = reader(read_obj)
    for row in csv_reader:
        products.append(row)
```

```python
items = set(sum(products, []))
```

```python
items
```

```python
s=list(items)
```

```python
df = pd.DataFrame(columns=s)
df
```

```python
encoder = TransactionEncoder()
transactions = encoder.fit(products).transform(products)
```

```python
transactions = transactions.astype('int')
```

```python
df = pd.DataFrame(transactions, columns=encoder.columns_)
```

```python
df.head()
```

```python
df.shape
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets.head(20)
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```
